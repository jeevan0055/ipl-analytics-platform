```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
plt.rcParams["figure.figsize"] = (12,6)

matches = pd.read_csv("../data/processed/matches_cleaned.csv")
```

```python
# ==========================================================
# MATCHES PLAYED PER SEASON
# ==========================================================

matches_per_season = (
    matches.groupby("season")
           .size()
           .reset_index(name="Matches")
)

display(matches_per_season)
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    data=matches_per_season,
    x="season",
    y="Matches"
)

plt.title("IPL Matches Played Per Season")
plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.xticks(rotation=45)

plt.show()
```

```python
# ==========================================================
# MOST SUCCESSFUL IPL TEAMS
# ==========================================================

team_wins = (
    matches["winner"]
    .value_counts()
    .reset_index()
)

team_wins.columns = ["Team","Wins"]

display(team_wins)
```

```python
plt.figure(figsize=(12,7))

sns.barplot(
    data=team_wins,
    y="Team",
    x="Wins"
)

plt.title("Most Successful IPL Teams")
plt.xlabel("Wins")
plt.ylabel("Team")

plt.show()
```

```python
# ==========================================================
# DOES WINNING THE TOSS HELP?
# ==========================================================

matches["toss_match_win"] = (
    matches["toss_winner"] == matches["winner"]
)

toss_impact = (
    matches["toss_match_win"]
    .value_counts()
    .reset_index()
)

toss_impact.columns = ["Won Toss & Match","Count"]

display(toss_impact)
```

```python
plt.figure(figsize=(6,6))

plt.pie(
    toss_impact["Count"],
    labels=["Yes","No"],
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Impact of Toss on Match Result")

plt.show()
```

```python
player_awards = (
    matches["player_of_match"]
    .value_counts()
    .head(15)
    .reset_index()
)

player_awards.columns = ["Player","Awards"]

display(player_awards)
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    data=player_awards,
    x="Awards",
    y="Player"
)

plt.title("Top 15 Player of the Match Award Winners")

plt.show()
```

```python
venues = (
    matches["venue"]
    .value_counts()
    .head(15)
    .reset_index()
)

venues.columns = ["Venue","Matches"]

display(venues)
```

```python
plt.figure(figsize=(12,7))

sns.barplot(
    data=venues,
    x="Matches",
    y="Venue"
)

plt.title("Top IPL Venues by Matches Hosted")

plt.show()
```

```python
matches_per_season.to_csv(
    "../data/processed/matches_per_season.csv",
    index=False
)

team_wins.to_csv(
    "../data/processed/team_wins.csv",
    index=False
)

player_awards.to_csv(
    "../data/processed/player_awards.csv",
    index=False
)

venues.to_csv(
    "../data/processed/venues.csv",
    index=False
)

print("EDA outputs saved successfully.")
```
