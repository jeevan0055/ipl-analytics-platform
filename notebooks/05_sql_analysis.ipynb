```python
import pandas as pd
import sqlite3

# Connect to SQLite database
conn = sqlite3.connect("../sql/ipl_analysis.db")

print("Connected to IPL Database Successfully!")
```

```python
query = """
SELECT name
FROM sqlite_master
WHERE type='table';
"""

tables = pd.read_sql(query, conn)

display(tables)
```

```python
query = """
SELECT COUNT(*) AS Total_Matches
FROM matches;
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT
    season,
    COUNT(*) AS Matches
FROM matches
GROUP BY season
ORDER BY season;
"""

season_matches = pd.read_sql(query, conn)

display(season_matches)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

plt.bar(
    season_matches["season"],
    season_matches["Matches"]
)

plt.title("Matches Per Season (SQL)")
plt.xticks(rotation=45)

plt.show()
```
