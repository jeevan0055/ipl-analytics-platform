```python
# ==========================================================
# DUPLICATE RECORDS
# ==========================================================

print("Duplicate rows in Matches :", matches.duplicated().sum())
print("Duplicate rows in Deliveries :", deliveries.duplicated().sum())
```

```python
# ==========================================================
# UNIQUE TEAM NAMES
# ==========================================================

teams = pd.concat([
    matches["team1"],
    matches["team2"],
    matches["winner"],
    matches["toss_winner"]
]).dropna().unique()

teams = sorted(teams)

print(f"Total unique team names: {len(teams)}\n")

for team in teams:
    print(team)
```

```python
# ==========================================================
# STANDARDIZE TEAM NAMES
# ==========================================================

team_name_mapping = {
    "Delhi Daredevils": "Delhi Capitals",
    "Kings XI Punjab": "Punjab Kings",
    "Royal Challengers Bangalore": "Royal Challengers Bengaluru",
    "Rising Pune Supergiants": "Rising Pune Supergiant"
}

team_columns = [
    "team1",
    "team2",
    "winner",
    "toss_winner"
]

for column in team_columns:
    matches[column] = matches[column].replace(team_name_mapping)

print("✅ Team names standardized successfully!")
```

```python
teams = pd.concat([
    matches["team1"],
    matches["team2"],
    matches["winner"],
    matches["toss_winner"]
]).dropna().unique()

teams = sorted(teams)

print(f"Total unique team names: {len(teams)}\n")

for team in teams:
    print(team)
```

```python
# ==========================================================
# MISSING VALUE PERCENTAGE
# ==========================================================

missing = matches.isnull().sum().to_frame(name="Missing Values")
missing["Percentage"] = (missing["Missing Values"] / len(matches) * 100).round(2)

missing = missing[missing["Missing Values"] > 0]

print("Missing Value Summary")
display(missing.sort_values("Percentage", ascending=False))
```

```python
# ==========================================================
# INVESTIGATE MISSING VALUES
# ==========================================================

matches[matches.isnull().any(axis=1)]
```

```python
# ==========================================================
# HANDLE MISSING VALUES
# ==========================================================

# City
matches["city"] = matches["city"].fillna("Unknown")

# Result Margin
matches["result_margin"] = matches["result_margin"].fillna(0)

# Winner
matches["winner"] = matches["winner"].fillna("No Result")

# Player of Match
matches["player_of_match"] = matches["player_of_match"].fillna("No Award")

# Target Runs
matches["target_runs"] = matches["target_runs"].fillna(0)

# Target Overs
matches["target_overs"] = matches["target_overs"].fillna(0)

print("Missing values handled successfully!")
```

```python
# ==========================================================
# VERIFY MISSING VALUES
# ==========================================================

print(matches.isnull().sum())
```

```python
# ==========================================================
# SAVE CLEANED DATA
# ==========================================================

matches.to_csv("../data/processed/matches_cleaned.csv", index=False)

print("Cleaned matches dataset saved successfully!")
```
